## Generate the Dataframe
Steps:
  * load in every data frame and make it unique by patient
    * For categorical, we are going to just select the mode for each group and set it as the label. 
    * numeric is already unique by patient
  * join every dataframe by subject_id
  * split the dataframe by clusters and create a list of dfs
    * whole df, cluster1, cluster2, ..., clusterN
  * For each of the dataframes:
    * if a column is missing > 50% of values then drop
    * impute the average value for all other missing data
    * save and return

```python
# generate our base patient dataframe
patient_df = genPatientDF(patient_dir, files, cluster_df, mortality)
patient_df.head()
```

```python
# split our dataframes into a dictionary of dataframes
# includes a df of each cluster and the total dataframe
def splitPatientDFs(patient_df):
    split_dfs = {val:pd.DataFrame(df).drop(columns = 'cluster') 
                 for val, df in patient_df.groupby('cluster', as_index=False)}
    split_dfs['total'] = patient_df.drop(columns = 'cluster')

    return split_dfs 

split_dfs = splitPatientDFs(patient_df)
split_dfs
```

```python
missing_percent = .5
for cluster, df in split_dfs.items():
    print(cluster)
    # process the data
    missing_df = dropMissings(df, missing_percent)
    imputed_df = imputeMissings(missing_df)
    
    # update the dictionary
    split_dfs[cluster] = imputed_df
```

```python
# now export dataframes 
for cluster, df in split_dfs.items():
    df.to_csv(output_dir + f"{cluster}_patient_df.csv", index = False)
```
